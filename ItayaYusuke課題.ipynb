```python
'''
#台形の面積
上辺：10cm
下辺：20cm
高さ：5cm

公式：台形の面積 =（上辺＋下辺）× 高さ ÷ 2
'''
top_side=10
bottom_side=20
height=5

area=(top_side+bottom_side)*height/2

print(f"{area}㎠")
```
