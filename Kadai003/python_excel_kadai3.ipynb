```python
import pandas as pd

# 2022 年間売上表ファイルを読み込む
df_2022 = pd.read_excel('2022_年間売上表.xlsx')

# 2023 年間売上表ファイルを読み込む
df_2023 = pd.read_excel('2023_年間売上表.xlsx')
# 2つのファイル
df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# 「商品」と「売上年」でグループ化して、金額を合計する（groupby）
df_summary = df_all.groupby(['商品', '売上年']).sum()

#  一旦、Excelファイルに書き出す
df_summary.to_excel('売上集計表.xlsx')

print(df_summary)

#  openpyxlで調整
import openpyxl
from openpyxl.styles import PatternFill

# 5. 保存したファイルをopenpyxlで読み込む
wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb.active

# 6. 1行目（ヘッダー）の色を変える設定
# 以前学んだ「色の塗りつぶし（PatternFill）」を使います
fill = PatternFill(patternType='solid', fgColor='F2F2F2') # 薄いグレー

# 1行目の全セルに対してループで色を塗る
# 1列目から、データが入っている最後の列まで処理します
for cell in ws[1]:
    cell.fill = fill

# 7. 最終保存
wb.save('売上集計表.xlsx')

```
