```python

""""
・変数varが、3の倍数の場合は「Fizz」を出力
・変数varが、5の倍数の場合は「Buzz」を出力
・変数varが、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力
・上記のどの場合にも該当しない場合は、変数varの値を出力

ただし、変数varは正の整数とします。
"""
#変数varが、3の倍数の場合は「Fizz」を出力
var =18  # ここに正の整数を入れます

# 1. 両方の倍数（15の倍数）を一番最初に！
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
# 2. 次に3の倍数
elif var % 3 == 0:
    print("Fizz")
# 3. 次に5の倍数
elif var % 5 == 0:
    print("Buzz")
# 4. どれでもない場合
else:
    print(var)
```
