```python
"""
課題の内容
要件に従って、社員ごとの売上データから各社員ごとの売上平均値と業績ランクを算出し、「業績.xlsx」を新規作成するプログラムを記述してください。

社員ごとの売上データは以下の通りです。

日付	社員名	売上	部門
2023-05-17	山田	100	メーカー
2023-05-18	佐藤	200	代理店
2023-05-19	鈴木	150	メーカー
2023-05-20	田中	300	商社
2023-05-21	高橋	250	代理店

「平均売上」列を作成し、全体の売上の平均を算出します。
「業績ランク」列を作成し、以下の条件分岐をして業績ランクを設定します。
関数「performance」を定義して業績ランクの設定をしてください。
業績ランク	条件
A	平均売上+50以上の場合
B	平均売上以上の場合
C	平均売上未満の場合
新規ファイル「業績.xlsx」に書き出します。
"""
!pip install pandas
```

```python
import pandas as pd

# データフレームの作成
df = pd.DataFrame({
    '日付':
      ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店'],
})

# 売上の平均を求めて新しい列を作成
df['平均売上'] = df['売上'].mean()

sales_average = df['平均売上'].mean()

# ランクで分岐して緊急度を求める関数「performance」を定義
def performance(sales):
  rank = ''
  if sales >=sales_average + 50 :
    rank = 'A'
  elif sales >= sales_average :
    rank = 'B'
  else:
    rank = 'C'
  return rank

# 「業績ランク」列を作成し、関数「performance」を適用して値を設定
df['業績ランク'] = df['売上'].apply(performance)

# Excelファイルを作成
writer = pd.ExcelWriter('業績.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
df.to_excel(writer, sheet_name='Sheet1', index=False)

# Excelファイルを閉じる
writer.close()
```
